```python
from tkinter import *
import tkinter as tk
import ipywidgets as widgets
from IPython.display import display, clear_output
from tkinter import *
from PIL import ImageTk, Image
import json
import warnings
from tkinter import ttk
from tkcalendar import Calendar,DateEntry
from tkinter.messagebox import showinfo
import os 
import PIL 
import glob
from datetime import datetime
import speech_recognition as sr  

#Creating the window woth tkinter.
window=Tk()
window.title("Fake News Tool")
window.geometry('800x430')
window.resizable(False,False)
window.config(background = "#213141")

#First frame to put the image.

frame1= Frame(window, width=100, highlightbackground='black', highlightthicknes=3)
frame1.grid(row=1,column=0,padx=0,pady=0, ipadx=0,ipady=0)

#Adding the image.
image = Image.open("OIP.jpg")
photo = ImageTk.PhotoImage(image)

img_label = Label(frame1,image=photo,width = "300", height = "270")
img_label.grid(row=1,column=0,padx=0,pady=0, ipadx=0,ipady=0)

#Second Frame to create the form and voice recognition button.

frame3= Frame(window, width=100, highlightbackground='black', highlightthicknes=3)
frame3.grid(row=1,column=1,padx=20,pady=20, ipadx=175,ipady=170)

username_label = Label(frame3,text = "Username(*):", bg = "#FFEEDD")
username_label.place(x =20, y = 20)

username = StringVar()
username_entry = Entry(frame3,textvariable = username, width = "40")
username_entry.place(x = 130, y = 20)

characteristic_label = Label(frame3,text = "Characteristic(*):", bg = "#FFEEDD")
characteristic_label.place(x = 20, y = 55)


n = tk.StringVar()
monthchoosen = ttk.Combobox(frame3, width = 27, 
                            textvariable = n)
  
# Adding combobox drop down list
monthchoosen['values'] = ('Politician', 
                          'Influencer',
                          'Journalist')
monthchoosen.place(x = 130, y = 55)
monthchoosen.current(0) 

cal_label = Label(frame3,text = "Pick a range date", bg = "#FFEEDD")
cal_label.place(x = 170, y = 90)

calini_label = Label(frame3,text = "Start date(*):", bg = "#FFEEDD")
calini_label.place(x = 70, y = 120)

calinidate=StringVar()
calini = DateEntry(frame3,textvariable = calinidate,width=30,bg="darkblue",fg="white",year=2022, month=4, day=1)
calini.place(x=20, y=145)

calend_label = Label(frame3,text = "End date(*):", bg = "#FFEEDD")
calend_label.place(x = 275, y = 120)

calenddate=StringVar()
calend = DateEntry(frame3,textvariable = calenddate, width=30,bg="darkblue",fg="white",year=2022)
calend.place(x=235, y=145)

tweets2_label = Label(frame3,text = "Number of tweets(*):", bg = "#FFEEDD")
tweets2_label.place(x = 20, y = 190)


var =IntVar()
var1 = IntVar()
def selection():
    choice = var.get()
    if choice == 1:
        m = '1'
        w2.place(x=1000, y=1000)
    elif choice == 2:
        m = '0'
        w2.place(x=260, y=200)
        pass
    return m 

Alltweets=Radiobutton(frame3, text='All default tweets', variable=var, value=1,command=selection)
Alltweets.place(x=160,y=180)
Alltweets.select()
Less100=Radiobutton(frame3, text='Less than 100', variable=var, value=2,command=selection)
Less100.place(x=160,y=210)


w2 = Scale(frame3, from_=1,to=99, variable = var1, orient=HORIZONTAL)
w2.set(50)
w2.pack()
w2.place(x=1000,y=1000)


def show_values():
    
    
    
    if (username_entry.get()!=""):
                  
        
        if(calend.get_date()>=calini.get_date()):
            window2 = tk.Toplevel()
            canvas = tk.Canvas(window2, height=150, width=400)
            window2.grab_set()
            canvas.pack()
        
            def windows_destroy():
                window2.destroy() 
                window.destroy()
                
            usernamew2 = Label(window2,text = f"This is the twitter account of {monthchoosen.get()}, {username_entry.get()}.", bg = "#FFEEDD")
            usernamew2.place(x = 20, y = 20)
            dateiniw2 = Label(window2,text = f"Twittes from {calini.get_date()}", bg = "#FFEEDD")
            dateiniw2.place(x = 20, y = 40)
            dateendw2 = Label(window2,text = f"Twittes till {calend.get_date()}", bg = "#FFEEDD")
            dateendw2.place(x = 20, y = 60)
            
            if(var.get()==1):
                
                tweetsw2 = Label(window2,text = f"Number of default twitts {100}", bg = "#FFEEDD")
                tweetsw2.place(x = 20, y = 80)
            else:
                
                tweetsw2 = Label(window2,text = f"Number of twitts {w2.get()}", bg = "#FFEEDD")
                tweetsw2.place(x = 20, y = 80)
                
            Back=Button(window2, text='Go back', command=lambda:window2.destroy())
            Back.pack()
            Back.place(x=20,y=120)
        
            CloseAll=Button(window2, text='Close all windows', command=windows_destroy)
            CloseAll.pack()
            CloseAll.place(x=100,y=120)
        else:
            window7 = tk.Toplevel()
            canvas = tk.Canvas(window7, height=200, width=230)
            window7.grab_set()
            canvas.pack()
            
            def windows_destroy5():
                window7.destroy() 
                window.destroy()
                
            img_label_empty = Label(window7, image= photo_empty)
            img_label_empty.pack()
            img_label_empty.place(x=60, y=20)
            ErrorDates = Label(window7,text = "Final date is lower than Initial date.", bg = "#FFEEDD")
            ErrorDates.place(x = 20, y = 80)
        
            Back_empty=Button(window7, text='Go back', command=lambda:window7.destroy())
            Back_empty.pack()
            Back_empty.place(x=20,y=120)
       
        
            CloseAll=Button(window7, text='Close all windows', command=windows_destroy5)
            CloseAll.pack()
            CloseAll.place(x=90,y=120)
        
        #print(f"This is the twitter account {monthchoosen.get()}, {username_entry.get()}.")
        #print(f"Twittes of {calini.get_date()}")
        #print(f"Twittes of {calend.get_date()}")
        #print(f"Number of twitts {w2.get()}")
        
        
    else:
        if(calend.get_date()>=calini.get_date()):       
            window3 = tk.Toplevel()
            canvas = tk.Canvas(window3, height=200, width=200)
            window3.grab_set()
            canvas.pack()
        
            def windows_destroy1():
                window3.destroy() 
                window.destroy()
         
            img_label_empty = Label(window3, image= photo_empty)
            img_label_empty.pack()
            img_label_empty.place(x=60, y=20)
            Empty = Label(window3,text = "Error, username is empty", bg = "#FFEEDD")
            Empty.place(x = 20, y = 80)
        
            Back_empty=Button(window3, text='Go back', command=lambda:window3.destroy())
            Back_empty.pack()
            Back_empty.place(x=20,y=120)
       
        
            CloseAll=Button(window3, text='Close all windows', command=windows_destroy1)
            CloseAll.pack()
            CloseAll.place(x=90,y=120)
        else:
            window8 = tk.Toplevel()
            canvas = tk.Canvas(window8, height=230, width=230)
            window8.grab_set()
            canvas.pack()
        
            def windows_destroy6():
                window8.destroy() 
                window.destroy()
         
            img_label_empty = Label(window8, image= photo_empty)
            img_label_empty.pack()
            img_label_empty.place(x=60, y=20)
            Empty = Label(window8,text = "Error, username is empty", bg = "#FFEEDD")
            Empty.place(x = 20, y = 80)
            ErrorDates = Label(window8,text = "And, final date is lower than Initial date.", bg = "#FFEEDD")
            ErrorDates.place(x = 20, y = 100)
        
            Back_empty=Button(window8, text='Go back', command=lambda:window8.destroy())
            Back_empty.pack()
            Back_empty.place(x=20,y=120)
       
        
            CloseAll=Button(window8, text='Close all windows', command=windows_destroy6)
            CloseAll.pack()
            CloseAll.place(x=90,y=120)
        
    

Send=Button(frame3, text='Send to search', command=show_values)
Send.pack()
Send.place(x=165,y=237)

Lines = Label(frame3,text = "---------------------------------------------------------------------------------", bg = "#FFEEDD")
Lines.place(x =20, y = 260)

def Catching_Audio():
    r=sr.Recognizer()
    with sr.Microphone() as source:
        audio=r.listen(source)
    
    try:
        
        txt=r.recognize_google(audio)
        txt1=str(txt)
        label = Label(frame3,text=txt,bg = "#FFEEDD")
        label.place(x = 1000, y = 1000)
        textfinal=label.cget("text")
        
        window4 = tk.Toplevel()
        canvas = tk.Canvas(window4, height=200, width=200)
        window4.grab_set()
        canvas.pack()
        
        def windows_destroy2():
            window4.destroy() 
            window.destroy()
        
        labelw4 = Label(window4,text = txt, bg = "#FFEEDD")
        labelw4.place(x = 20, y = 80)
    
        
        Back_w4=Button(window4, text='Go back', command=lambda:window4.destroy())
        Back_w4.pack()
        Back_w4.place(x=20,y=120)
       
        
        CloseAllw4=Button(window4, text='Close all windows', command=windows_destroy2)
        CloseAllw4.pack()
        CloseAllw4.place(x=90,y=120)
        
        return texto.set(txt1)
        
    except sr.UnknownValueError:
        window5 = tk.Toplevel()
        canvas = tk.Canvas(window5, height=200, width=200)
        window5.grab_set()
        canvas.pack()
        
        def windows_destroy3():
            window5.destroy() 
            window.destroy()
            
        labelw5 = Label(window5,text="Could not understand audio, try again",bg = "#FFEEDD")
        labelw5.place(x = 20, y = 80)
        
        Back_w5=Button(window5, text='Go back', command=lambda:window5.destroy())
        Back_w5.pack()
        Back_w5.place(x=20,y=120)
       
        
        CloseAllw5=Button(window5, text='Close all windows', command=windows_destroy3)
        CloseAllw5.pack()
        CloseAllw5.place(x=90,y=120)
        
        
    except sr.RequestError as e:
        window6 = tk.Toplevel()
        canvas = tk.Canvas(window6, height=200, width=200)
        window6.grab_set()
        canvas.pack()
        
        def windows_destroy4():
            window6.destroy() 
            window.destroy()
            
        message=f"Could not request results; {0}, try again".format(e)
        labelw6 = Label(window6,text=message,bg = "#FFEEDD")
        labelw6.place(x = 200, y = 300)
        
        Back_w6=Button(window6, text='Go back', command=lambda:window6.destroy())
        Back_w6.pack()
        Back_w6.place(x=20,y=120)
       
        
        CloseAllw6=Button(window6, text='Close all windows', command=windows_destroy4)
        CloseAllw6.pack()
        CloseAllw6.place(x=90,y=120)
       
#Image to use in the small window.
image_empty = Image.open("empty.jfif")
resized_image_empty = image_empty.resize((70,70))
photo_empty = ImageTk.PhotoImage(resized_image_empty)


#Image to use in the next button.
image1 = Image.open("download.jfif")
resized_image = image1.resize((35,35))
photo1 = ImageTk.PhotoImage(resized_image)

   
#Variables to use in Catching_Audio
texto=tk.StringVar()

Audio=Button(frame3, text='Press and Speak', image = photo1, command=Catching_Audio, compound = LEFT)
Audio.pack()
Audio.place(x=20,y=284)

Speak_loud = Label(frame3,text = "If you do not know the twitter account,", bg = "#FFEEDD")
Speak_loud.place(x =200, y = 284)
Speak_loud1 = Label(frame3,text = "speak out the name of the country.     ", bg = "#FFEEDD")
Speak_loud1.place(x =200, y = 303)


FinalClose=Button(window, text='Close Window', command=lambda:window.destroy(), compound = LEFT)
FinalClose.place(x=300,y=380)

window.mainloop()
```
